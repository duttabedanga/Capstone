```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException  #importing exception, since we are using dynamic page/website where the elements in page might be relocated or added or value get change 


driver =webdriver.Chrome()


'''AMERICA AND CANADA'''

#opening the website on automated chrome browser (city- Austin TX)

driver.get("https://www.homes.com/austin-tx/homes-for-rent/")

```

```python
#Empty list for store 
Address=[]
Rent=[]
Details=[]
Bed=[]
bathroom=[]
Area=[]
Amenities=[]
URL=[]

#till 5 pages 
start=0
end=5


for page in range(start,end): #for loop for scrapping 5 page
    #Scraping URL
    for i in driver.find_elements(By.XPATH,'//div[@class="for-rent-content-container"]/a'):
        URL.append(i.get_attribute('href')) #to retrive any link we use get_attribute 
    

    # Wait for next button and click
    next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[@class="next text-only "]'))) #to scrape data from next page (here we can use 'find _elements' instead of 'find_element', so that no need to write 'next_button.click()' command in next line)
    next_button.click()
    time.sleep(3)
    
len(URL)
    
```

```python
for i in URL:       #iterating each and every URL to fetch full job details ( loop for everything in the list)
    driver.get(i)   # taking each URL one by one as an input
    time.sleep(2)
    
    #Scraping Address 
    try:
        location = driver.find_element(By.XPATH,'//div[@class="property-info-address"]')
        Address.append(location.text)
    
    except NoSuchElementException:
        Address.append('-')
        
    #Scraping Rental_price
    try:
        Rental_price = driver.find_element(By.XPATH,'//span[@class="property-info-price"]')
        Rent.append(Rental_price.text)
    
    except NoSuchElementException:
        Rent.append('-')
    
    
  
    #Scraping Number of Bed
    try:
        no_of_bed = driver.find_element(By.XPATH,'//span[@class="property-info-feature beds"]')
        Bed.append(no_of_bed.text)
    
    except NoSuchElementException:
        Bed.append('-')
        
        
        
     #Scraping Number of bathroom
    try:
        no_of_bathroom = driver.find_element(By.XPATH,'//span[@class="property-info-feature"]')
        bathroom.append(no_of_bathroom.text)
    
    except NoSuchElementException:
        bathroom.append('-')
        
        
  
    #Scraping Area
    try:
        space = driver.find_element(By.XPATH,'//span[@class="property-info-feature sqft"]')
        Area.append(space.text)
    
    except NoSuchElementException:
        Area.append('-')
        
        
    # Scraping Details
    try:
        all_Details = driver.find_elements(By.XPATH, '//span[@class="highlight-value"]')
        Details.append([i.text for i in all_Details])
        
    except NoSuchElementException:
        Details.append('-')
        
    
    
    
    #Scraping Amenities
    try:
        facility = driver.find_elements(By.XPATH,'//div[@class="subcategory-container"]')
        Amenities.append([i.text for i in facility]) # Iterate over the list and extract text from each element
            
            
    except NoSuchElementException:
        Amenities.append('-')
    
    
        
        
len(Address),len(Rent),len(Bed),len(bathroom),len(Area),len(Details),len(Amenities)         
```

```python
#Finally create a dataframe of the scraped data.
df1=pd.DataFrame({'Address':Address , 'Rent':Rent , 'Bed':Bed, 'bathroom':bathroom, 'Area':Area , 'Details':Details ,'Amenities':Amenities})
df1
```

```python
# Save to CSV
df1.to_csv("AMERICA AND CANADA (Austin TX).csv", index=False)  #file got saved in 'C:\Users\riki' location
```

```python
for i in URL:       #iterating each and every URL to fetch full job details ( loop for everything in the list)
    driver.get(i)   # taking each URL one by one as an input
    time.sleep(2)
    
    #Scraping Address 
    try:
        location = driver.find_element(By.XPATH,'//div[@class="property-info-address"]')
        Address.append(location.text)
    
    except NoSuchElementException:
        Address.append('-')
        
    #Scraping Rental_price
    try:
        Rental_price = driver.find_element(By.XPATH,'//span[@class="property-info-price"]')
        Rent.append(Rental_price.text)
    
    except NoSuchElementException:
        Rent.append('-')
    
    
  
    #Scraping Number of Bed
    try:
        no_of_bed = driver.find_element(By.XPATH,'//span[@class="property-info-feature beds"]')
        Bed.append(no_of_bed.text)
    
    except NoSuchElementException:
        Bed.append('-')
        
        
        
     #Scraping Number of bathroom
    try:
        no_of_bathroom = driver.find_element(By.XPATH,'//span[@class="property-info-feature"]')
        bathroom.append(no_of_bathroom.text)
    
    except NoSuchElementException:
        bathroom.append('-')
        
        
  
    #Scraping Area
    try:
        space = driver.find_element(By.XPATH,'//span[@class="property-info-feature sqft"]')
        Area.append(space.text)
    
    except NoSuchElementException:
        Area.append('-')
        
        
    # Scraping Details
    try:
        all_Details = driver.find_elements(By.XPATH, '//span[@class="highlight-value"]')
        Details.append([i.text for i in all_Details])
        
    except NoSuchElementException:
        Details.append('-')
        
    
    
    
    #Scraping Amenities
    try:
        facility = driver.find_elements(By.XPATH,'//div[@class="subcategory-container"]')
        Amenities.append([i.text for i in facility]) # Iterate over the list and extract text from each element
            
            
    except NoSuchElementException:
        Amenities.append('-')
    
    
        
        
len(Address),len(Rent),len(Bed),len(bathroom),len(Area),len(Details),len(Amenities)         
```

```python
#Finally create a dataframe of the scraped data.
df20=pd.DataFrame({'Address':Address , 'Rent':Rent , 'Bed':Bed, 'bathroom':bathroom, 'Area':Area , 'Details':Details ,'Amenities':Amenities})
df20
```

```python
import pandas as pd

df_rate = pd.DataFrame(df)

# Remove 'per month' from the 'Rent' column
df_rate['Rent'] = df_rate['Rent'].str.replace('per month', '')

# Display the updated DataFrame
df_rate
```

```python
import pandas as pd
import numpy as np

# Replace empty values '[]' with None
df_rate.replace('[]', None, inplace=True)

# Replace '-' with None
df_rate.replace('-', None, inplace=True)

df_rate
```

```python
import pandas as pd
import numpy as np

# Replacing 'None' in 'Rent' column with random values
def generate_random_rent(x):
    if x is None:
        return '$' + str(np.random.randint(1000, 5000))
    else:
        return x

# Applying the function to the 'Rent' column
df_rate['Rent'] = df_rate['Rent'].apply(generate_random_rent)


df_rate
```

```python
import pandas as pd


new_df = pd.DataFrame(df_rate)

# Defining the conversion rates
conversion_rates = {'USD': 83.12, 'Euro': 90.73, 'Swiss Franc': 95.73, 'Kuwaiti Dinar': 270.04, 'Dirham': 22.63}

def convert_to_inr(row):
    # Extracting the currency symbol and amount
    currency = row['Currency']
    amount = float(row['Amount'].replace(',', ''))

    
    # Convert to INR based on conversion rates
    if currency == '$':
        return round(amount * conversion_rates['USD'])
    elif currency == 'Euro':
        return round(amount * conversion_rates['Euro'])
    elif currency == 'Franc':
        return round(amount * conversion_rates['Swiss Franc'])
    elif currency == 'Dinar':
        return round(amount * conversion_rates['Kuwaiti Dinar'])
    elif currency == 'Dirham':
        return round(amount * conversion_rates['Dirham'])

    
# Extracting currency and amount into separate columns
df_rate[['Currency', 'Amount']] = df_rate['Rent'].str.extract('([^\d]+)(\d{1,3},?\d{0,3})', expand=True)

    
# Removing commas from 'Rent' column
df_rate['Rent'] = df_rate['Rent'].str.replace(',', '')
    
# Applying the conversion function to 'Rent' column
new_df['Rent_INR'] = df_rate.apply(convert_to_inr, axis=1)

# Drop the 'Rent' column
new_df = new_df.drop('Rent', axis=1)

# Updated DataFrame
new_df
```

```python
import pandas as pd

# Filling empty spaces if any in 'bathroom', 'Bed', 'Area' columns with  0
new_df['bathroom'].fillna(0, inplace=True)
new_df['Bed'].fillna(0, inplace=True)
new_df['Area'].fillna(0, inplace=True)

# Filling with 'Not available'
new_df['Details'].fillna('Not available', inplace=True)
new_df['Amenities'].fillna('Not available', inplace=True)
```
