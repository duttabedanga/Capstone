```python
'''Make the table in  csv format'''
#1) table1.csv : Sno,Address,City,State code,Country, Rental Price,deposit

# Adding 'Sno' and 'Dposit' column in the data frame

new_df['Sno'] = range(1, len(new_df) + 1)
new_df['Deposit'] = 0

new_df
```

```python
# Shifting 'Sno' column to the first position
new_df.insert(0, 'Sno', new_df.pop('Sno'))


new_df
```

```python
import pandas as pd

# Adding 'State code' as different column from the 'Address' column
new_df['State code'] = new_df['Address'].str.extract(r'(\d{5})')

new_df
```

```python
# Rename a specific column
new_df = new_df.rename(columns={'County': 'Country'})
new_df
```

```python
#2) table2.csv : Sno, no of bed, no.of bathroom , Area ,pets allowed

def transform_bathroom(x):
    if x.split("\n")[0].replace(".", "").isnumeric():
        return float(x.split("\n")[0])
    else: 
        return 0 
    
    
# Extracting numeric values from 'Bed' column
new_df['Bed'] = new_df['Bed'].astype(str).str.extract('(\d+)', expand=False).astype(float)

# Extracting numeric values from 'Bathroom' column
new_df['bathroom'] = new_df['bathroom'].apply(transform_bathroom)


new_df
```

```python

def pet_policy_check(x):
    if "pets" in x.lower() or "dog" in x.lower() or "cat" in x.lower():
        return True
    else:
        return False

# Extracting pets_allowed from 'Amenities' column
new_df['pets_allowed'] = new_df['Amenities'].apply(pet_policy_check)


new_df
```

```python
table2 = new_df[['Sno', 'Bed', 'bathroom', 'Area', 'pets_allowed']]
table2
```

```python
# Save to CSV
table2.to_csv("table2.csv", index=False)  #file got saved in 'C:\Users\riki' location
```

```python
#3) table3.csv : Sno ,Washer/Dryer , AC ,Parking, Fireplace, Dishwasher, Hardwood floors, Roofdeck ,storage


def Washer_Dryer(x):
    if "washer" in x.lower() or "dryer" in x.lower():
        return True
    else:
        return False

# Extracting Washer or Dryer from 'Amenities' column
new_df['Washer/Dryer'] = new_df['Amenities'].apply(Washer_Dryer)



def Parking(x):
    if "parking" in x.lower():
        return True
    else:
        return False

# Extracting Parking from 'Amenities' or 'Details' column
new_df['Parking'] = new_df['Amenities'].combine_first(new_df['Details']).apply(Parking) #'combine_first' to handle cases where one of them is null



def Fireplace(x):
    if "fireplace" in x.lower():
        return True
    else:
        return False

# Extracting Fireplace from 'Amenities' or 'Details' column
new_df['Fireplace'] = new_df['Amenities'].combine_first(new_df['Details']).apply(Fireplace) #'combine_first' to handle cases where one of them is null



def AC(x):
    if "air condition" in x.lower():
        return True
    else:
        return False

# Extracting air condition from 'Amenities' or 'Details' column
new_df['AC'] = new_df['Amenities'].combine_first(new_df['Details']).apply(AC) 



def Dishwasher(x):
    if "dishwasher" in x.lower():
        return True
    else:
        return False

# Extracting Dishwasher from 'Amenities' or 'Details' column
new_df['Dishwasher'] = new_df['Amenities'].combine_first(new_df['Details']).apply(Dishwasher) 




def Hardwood_floors(x):
    if "hardwood floor" in x.lower():
        return True
    else:
        return False

# Extracting Hardwood_floors from 'Amenities' or 'Details' column
new_df['Hardwood_floors'] = new_df['Amenities'].combine_first(new_df['Details']).apply(Hardwood_floors) 


def Roofdeck(x):
    if "roof" in x.lower():
        return True
    else:
        return False

# Extracting Roofdeck from 'Amenities' or 'Details' column
new_df['Roofdeck'] = new_df['Amenities'].combine_first(new_df['Details']).apply(Roofdeck )



def storage(x):
    if "storage" in x.lower():
        return True
    else:
        return False

# Extracting storage from 'Amenities' or 'Details' column
new_df['storage'] = new_df['Amenities'].combine_first(new_df['Details']).apply(storage)



new_df
```
