```python
#Q.Write the complete Machine learning code to make predictions of rental price and area .Use appropriate models on their label basis. Remember you need to make 2 different predictions: rental price  and area   .
#Apply all the best techniques of scaling ,hyperparameter tuning , avoid underfitting or overfitting (bias/variance)
#At the end save the best model and convey on which basis you have chosen that model.  


import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from joblib import dump


# Droping unnecessary columns
columns_to_drop = ['Address', 'State code', 'Country']
merged_df = merged_df.drop(columns=columns_to_drop)
```

```python
# Encode boolean columns using LabelEncoder
bool_columns = ['pets_allowed', 'Washer/Dryer', 'AC', 'Parking', 'Fireplace', 'Dishwasher', 'Hardwood_floors', 'Roofdeck', 'storage']
le = LabelEncoder()

for col in bool_columns:
    merged_df[col] = le.fit_transform(merged_df[col])
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder

# Extract categorical columns
categorical_columns = ['City']

# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
categorical_encoded = pd.DataFrame(encoder.fit_transform(merged_df[categorical_columns]), columns=encoder.get_feature_names(categorical_columns))
merged_df = pd.concat([merged_df, categorical_encoded], axis=1)
merged_df = merged_df.drop(columns=categorical_columns)
```

```python
merged_df
```

```python
# Spliting the data into features (X) and labels (y) for both predictions
X_rental = merged_df.drop(['Rent_INR'], axis=1)
y_rental = merged_df['Rent_INR']

X_area = merged_df.drop(['Area'], axis=1)
y_area = merged_df['Area']

# Split the data into training and testing sets
X_rental_train, X_rental_test, y_rental_train, y_rental_test = train_test_split(X_rental, y_rental, test_size=0.2, random_state=42)
X_area_train, X_area_test, y_area_train, y_area_test = train_test_split(X_area, y_area, test_size=0.2, random_state=42)

# Standardize features
scaler_rental = StandardScaler()
X_rental_train_scaled = scaler_rental.fit_transform(X_rental_train)
X_rental_test_scaled = scaler_rental.transform(X_rental_test)

scaler_area = StandardScaler()
X_area_train_scaled = scaler_area.fit_transform(X_area_train)
X_area_test_scaled = scaler_area.transform(X_area_test)
```

```python
# Hyperparameter tuning using GridSearchCV for rental prediction
param_grid_rental = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
rf_rental = RandomForestRegressor(random_state=42)
grid_search_rental = GridSearchCV(rf_rental, param_grid_rental, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_rental.fit(X_rental_train_scaled, y_rental_train)
```

```python
# Getting the best hyperparameters
best_params_rental = grid_search_rental.best_params_
```

```python
# Training the best model for rental prediction
best_rf_rental = RandomForestRegressor(random_state=42, **best_params_rental)
best_rf_rental.fit(X_rental_train_scaled, y_rental_train)
```

```python
# Hyperparameter tuning using GridSearchCV for area prediction
param_grid_area = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
rf_area = RandomForestRegressor(random_state=42)
grid_search_area = GridSearchCV(rf_area, param_grid_area, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_area.fit(X_area_train_scaled, y_area_train)
```

```python
# Get the best hyperparameters
best_params_area = grid_search_area.best_params_
```

```python
# Training the best model for area prediction
best_rf_area = RandomForestRegressor(random_state=42, **best_params_area)
best_rf_area.fit(X_area_train_scaled, y_area_train)
```

```python
# Saving the best models
dump(best_rf_rental, 'best_rf_rental_model.joblib')
dump(best_rf_area, 'best_rf_area_model.joblib')
```

```python
# Predictions on test sets
y_rental_pred = best_rf_rental.predict(X_rental_test_scaled)
y_area_pred = best_rf_area.predict(X_area_test_scaled)

```

```python
# Evaluating the models
rental_mse = mean_squared_error(y_rental_test, y_rental_pred)
area_mse = mean_squared_error(y_area_test, y_area_pred)
```

```python
print(f"Best Model for Rental Price - Mean Squared Error: {rental_mse}")
print(f"Best Model for Area - Mean Squared Error: {area_mse}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_rental_test, y=y_rental_test - y_rental_pred)
plt.title('Residuals for Rental Price Prediction')
plt.xlabel('Actual Rental Price')
plt.ylabel('Residuals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_area_test, y=y_area_test - y_area_pred)
plt.title('Residuals for Area Prediction')
plt.xlabel('Actual Area')
plt.ylabel('Residuals')
plt.show()
```
